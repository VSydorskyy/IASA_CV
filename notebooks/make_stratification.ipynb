```python
import pandas as pd
import numpy as np

from sklearn.model_selection import StratifiedKFold
```

## Read data

```python
df = pd.read_csv('../data/heart.csv')

df
```

## Stratification by target

```python
skf = StratifiedKFold(
    n_splits=5, 
    shuffle=True, 
    random_state=42
)

folds = list(skf.split(df, df['target']))
```

```python
for idx, f in enumerate(folds):
    print(f"Fold {idx}")
    print(f'Train len: {len(f[0])}\nTest len: {len(f[1])}')
    print('Target distribution')
    print('Train : ')
    print(df.iloc[f[0]].target.value_counts(normalize=True))
    print('Test : ')
    print(df.iloc[f[1]].target.value_counts(normalize=True))
```

```python
np.save('../data/5folds_bytarget.npy', folds)
```
